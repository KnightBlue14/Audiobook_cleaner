```python
def hrs(time):
  numbers = re.findall(r'\d+', time)
  hours = int(numbers[0]) if numbers and 'hrs' in time else 0
  minutes = int(numbers[1]) if len(numbers) > 1 and 'mins' in time else 0
  total_minutes = hours * 60 + minutes
  return total_minutes
df['timetest'] = df.time.apply(hrs)
```

```python
df.stars = df.stars.str.replace('0 ratings','Not rated yet')
```

```python
#Function to take the initial string, split it about the word 'stars', then convert it into an integer rating, as well as replacing the phrase 'Not rated yet' with 0, so it can also be converted
def star_call(string):
  string = string.replace('Not rated yet','0 out of 5 stars0 ratings')
  string = re.split('(stars)',string)
  string = string[0]
  string = string.replace(' out of ',',').split(',')
  string = ''.join(str(i) for i in string[0])
  string = float(string)
  return string

df['rating'] = df.stars.apply(star_call)
```

```python
#del df['rating']
```

```python
#As earlier, but for the number of ratings
def rating_call(string):
  string = string.replace('Not rated yet','0 out of 5 stars0 ratings')
  string = re.split('(stars)',string)
  string = string[-1]
  string = string.replace(' ratings','')
  string = string.replace(' rating','')
  string = string.replace(',','')
  string = int(float(string))
  return string

df['number_of_ratings'] = df.stars.apply(rating_call)
```

```python
#Converts datetime
df.releasedate = pd.to_datetime(df.releasedate, format='%d-%m-%y')
```

```python
#Renames releasedate column
df.releasedate = df.rename(columns = {'releasedate': 'release_date'},inplace = True)
```

```python
#Removes commas, replaces Free with 0, then converts to float. Actual values unchanged, as the unit is unknown
df.price = df.price.str.replace(',', '')
df.price = df.price.str.replace('Free', '0.00')
df.price = pd.to_numeric(df.price)
```

```python
#Author and Narrator fixes (Just use the bottom one)
```

```python
#Removes Writtenby, Narrated by, and fullcast labels
df.author = df.author.str.replace('Writtenby:','')
df.narrator = df.narrator.str.replace('Narratedby:','')
df.narrator = df.narrator.str.replace('fullcast,','')
```

```python
#Adds spaces before Capitals letters, then removes whitespace
df.author = df.author.str.replace( r"([A-Z])", r" \1").str.strip()
df.narrator = df.narrator.str.replace( r"([A-Z])", r" \1").str.strip()
```

```python
#Rejoins names beginning with Mc i.e. McManus
df.author = df.narrator.str.replace('Mc ', 'Mc')
df.narrator = df.narrator.str.replace('Mc ', 'Mc')
```

```python
#Applies earlier functions
def fix_name(name):
  name = name.replace('Writtenby:','')
  name = name.replace('Narratedby:','')
  name = name.replace('fullcast,','')
  name = re.sub( r"([A-Z])", r" \1",name).strip()
  name = name.replace('Mc ', 'Mc')
  return name

df.author = df.author.apply(fix_name)
df.narrator = df.narrator.apply(fix_name)
```
